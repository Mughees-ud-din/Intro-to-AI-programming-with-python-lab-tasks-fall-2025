Task 2:	File Handling: Integer number and string  
Please read the attached CSV file (you can easily create it using the CSV module). Once you have read the rows, add 2 to each row and print the result.  
File name: numbers.csv  


```python
import csv
numbers=list(range(1,11)) #writing a csv file with numbers from 1 to 10
with open(r"numbers.csv",mode='w',newline='') as file:
    writer=csv.writer(file)
    for number in numbers:
     writer.writerow([number])
# now reading the created file and adding two in each row
with open(r'numbers.csv') as f: 
    reader = csv.reader(f)
    numbers = [float(row[0])+2 for row in reader]
    #printing the result
    print(numbers) 
    #counting number of integers
    count = sum(num.is_integer() for num in numbers)
    print("Total integers:", count)
```

Task 3: You are given a CSV file named integers.csv that contains one number per row (some are integers, and some are decimal numbers). Write a Python program that: 
1.	Reads all numbers from the CSV file. 
2.	Converts them into floating-point numbers. iii. 	Counts how many of them are whole numbers (integers). iv. 	Prints the total count of integers.
 
Note: You must not use any if statements in your program. Instead, use expressions like .is_integer() and built-in functions like sum(). 


```python
import csv
num=[]
with open(r'integers.csv') as f:
    reader = csv.reader(f)
    numbers = [row[0] for row in reader]
    print(numbers)
    for i in numbers:
     n=float(i.strip())
     num.append(n)
    print(num)
count=sum(n.is_integer() for n in num)
print("total integers are",count)
```

Task 6: Find Total Impedance  
Add the three impedances.  
Print real part (resistive) and imaginary part (reactive).  
Print magnitude and phase angle (use abs() and cmath.phase()). 


```python
import cmath

# Given impedances
Z1 = complex(30, 40)   # 30 + j40 Œ©
Z2 = complex(10, -20)  # 10 - j20 Œ©
Z3 = complex(50, 10)   # 50 + j10 Œ©

# iv. Add the three impedances
Z_total = Z1 + Z2 + Z3

# v. Print real and imaginary parts
print("Resistive (Real) part:", Z_total.real, "Œ©")
print("Reactive (Imaginary) part:", Z_total.imag, "Œ©")

# vi. Magnitude and phase angle
magnitude = abs(Z_total)
phase_rad = cmath.phase(Z_total)
phase_deg = phase_rad * 180 / 3.14159

print("Magnitude of total impedance:", magnitude, "Œ©")
print("Phase angle of total impedance:", phase_deg, "¬∞")
```

Task 7: Compute Circuit Current  
i.	Assume AC voltage source V=120 0¬∞V  (so in Python V = complex(120, 0)).  
ii.	Compute current:  
iii.	Print current magnitude and phase. 


```python
import cmath

# Given AC voltage source
V = complex(120, 0)  # 120‚à†0¬∞

# Example impedance (Z = 30 + j40 Œ©)
Z = complex(30, 40)

# Compute current
I = V / Z

# Compute magnitude and phase
magnitude = abs(I)
phase_rad = cmath.phase(I)
phase_deg = phase_rad * 180 / 3.14159

# Print results
print(f"Voltage (V): {V}")
print(f"Impedance (Z): {Z}")
print(f"Current (I): {I}")
print(f"Magnitude of current: {magnitude:.2f} A  ")
print(f"Phase angle of current: {phase_deg:.2f}¬∞")
```
